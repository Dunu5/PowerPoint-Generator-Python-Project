```python
from transformers import pipeline
import csv
import pandas as pd
import timeit

# Function for original sentiment analysis code
def original_sentiment_analysis(texts):
    classifier = pipeline('sentiment-analysis')
    results = classifier(texts)
    for text, result in zip(texts, results):
        print(f"Text: {text}\nSentiment: {result['label']}, Score: {round(result['score'], 4)}")

# Function for modified sentiment analysis code
def modified_sentiment_analysis(texts, output_file):
    classifier = pipeline('sentiment-analysis')
    results = classifier(texts)
    with open(output_file, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Text", "Label", "Score"])
        for text, result in zip(texts, results):
            writer.writerow([text, result['label'], round(result['score'], 4)])
    df = pd.read_csv(output_file)
    print(df)

# Example texts
texts = ["I love this!", "I hate this!"] * 5  # 10 texts
output_file = "results.csv"
```

```python
# Benchmark original code
original_time = timeit.timeit(lambda: original_sentiment_analysis(texts), number=1)

# Benchmark modified code
modified_time = timeit.timeit(lambda: modified_sentiment_analysis(texts, output_file), number=1)

print(f"Time taken by original code: {original_time} seconds")
print(f"Time taken by modified code: {modified_time} seconds")
```

```python
from transformers import pipeline
import csv
import pandas as pd
from memory_profiler import memory_usage

# Function for original sentiment analysis code
def original_sentiment_analysis(texts):
    classifier = pipeline('sentiment-analysis')
    results = classifier(texts)
    for text, result in zip(texts, results):
        print(f"Text: {text}\nSentiment: {result['label']}, Score: {round(result['score'], 4)}")

# Function for modified sentiment analysis code
def modified_sentiment_analysis(texts, output_file):
    classifier = pipeline('sentiment-analysis')
    results = classifier(texts)
    with open(output_file, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Text", "Label", "Score"])
        for text, result in zip(texts, results):
            writer.writerow([text, result['label'], round(result['score'], 4)])
    df = pd.read_csv(output_file)
    print(df)

# Example texts
texts = ["I love this!", "I hate this!"] * 5  # 10 texts
output_file = "results.csv"
```

```python
# Measure memory usage for original code
original_memory_usage = memory_usage((original_sentiment_analysis, (texts,)), interval=0.1)

# Measure memory usage for modified code
modified_memory_usage = memory_usage((modified_sentiment_analysis, (texts, output_file)), interval=0.1)

print(f"Memory usage by original code: {max(original_memory_usage)} MiB")
print(f"Memory usage by modified code: {max(modified_memory_usage)} MiB")
```

```python
# Example larger texts
large_texts = ["I love this!", "I hate this!"] * 1000  # 2000 texts

# Benchmark original code with larger dataset
large_original_time = timeit.timeit(lambda: original_sentiment_analysis(large_texts), number=1)
large_original_memory_usage = memory_usage((original_sentiment_analysis, (large_texts,)), interval=0.1)

# Benchmark modified code with larger dataset
large_modified_time = timeit.timeit(lambda: modified_sentiment_analysis(large_texts, output_file), number=1)
large_modified_memory_usage = memory_usage((modified_sentiment_analysis, (large_texts, output_file)), interval=0.1)

print(f"Time taken by original code (large dataset): {large_original_time} seconds")
print(f"Memory usage by original code (large dataset): {max(large_original_memory_usage)} MiB")
print(f"Time taken by modified code (large dataset): {large_modified_time} seconds")
print(f"Memory usage by modified code (large dataset): {max(large_modified_memory_usage)} MiB")
```
